```python
print("Welcome to ServiceHive Assistant ")
print("Ask me anything about our product!")
print("Type 'exit' to stop.\n")
```

```python
product_info = {
    "features": "ServiceHive is a SaaS platform that helps businesses automate customer support and manage leads efficiently.",
    "pricing": "Our pricing starts from $29 per month for small teams.",
    "demo": "You can request a free demo of ServiceHive anytime.",
    "support": "We provide 24/7 customer support with AI-powered assistance."
}
```

```python
def detect_intent(user_input):
    high_intent_keywords = ["buy", "purchase", "demo", "contact", "trial", "sign up"]

    for word in high_intent_keywords:
        if word in user_input.lower():
            return "high_intent"
    return "general_query"
```

```python
def answer_query(user_input):
    if "feature" in user_input.lower():
        return product_info["features"]
    elif "price" in user_input.lower() or "cost" in user_input.lower():
        return product_info["pricing"]
    elif "demo" in user_input.lower():
        return product_info["demo"]
    elif "support" in user_input.lower():
        return product_info["support"]
    else:
        return "I can help with features, pricing, demo, or support. Please ask related questions."
```

```python
import pandas as pd

def save_lead(name, email, company):
    df = pd.DataFrame([[name, email, company]], columns=["Name", "Email", "Company"])
    df.to_csv("leads.csv", mode='a', header=not pd.io.common.file_exists("leads.csv"), index=False)
    print(" Lead saved successfully!")
```

```python
while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Assistant: Thank you for chatting with ServiceHive!")
        break

    intent = detect_intent(user_input)

    if intent == "high_intent":
        print("Assistant: I see you're interested! Please share your details.")
        name = input("Your Name: ")
        email = input("Your Email: ")
        company = input("Your Company: ")
        save_lead(name, email, company)
        print("Assistant: Thanks! Our team will contact you soon.")

    else:
        response = answer_query(user_input)
        print("Assistant:", response)
```
